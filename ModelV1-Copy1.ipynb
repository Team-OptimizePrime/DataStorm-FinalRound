```python
final_2017 = pd.read_csv("final2017.csv")
final_2018 = pd.read_csv("final2018.csv")
final_2019 = pd.read_csv("final2019.csv")

cust_set_2017 = set(final_2017["customer_id"])
cust_set_2018 = set(final_2018["customer_id"])
cust_set_2019 = set(final_2019["customer_id"])
no_noise_cus_2017 = []
no_noise_cus_2018 = []
no_noise_cus_2019 = []

def data_clean(d,s,o):
    for i in d:
        #using number of months visited
        months = ["jan_4","feb_4","mar_4","apr_4","may_4","jun_4","jul_4","aug_4","sep_4","oct_4", "nov_4", "dec_4" ]
        tot_num_visits = 0
        for mnths in months:
            frame = pd.DataFrame(s.loc[s["customer_id"] == i][mnths])
            if frame.iloc[0][mnths] != 0.0:
                tot_num_visits += 1
        if tot_num_visits < 3: #threshold for months
            o.append(i)
```

```python
d = cust_set_2017
s = final_2017
o = no_noise_cus_2017
data_clean(d,s,o)
```

```python
d = cust_set_2018
s = final_2018
o2 = no_noise_cus_2018
data_clean(d,s,o2)
```

```python
d = cust_set_2019
s = final_2019
o3 = no_noise_cus_2019
data_clean(d,s,o3)
```

```python
def intersection(lst1, lst2): 
    return list(set(lst1) & set(lst2)) 
```

```python
remove = intersection(o, intersection(o3,o2))
len(remove)
```

```python
def Union(lst1, lst2): 
    final_list = list(set(lst1) | set(lst2)) 
    return final_list 
```
