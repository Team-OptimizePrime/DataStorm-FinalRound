```python
import pandas as pd
import numpy as np
```

```python
customer_data = pd.read_csv("customer_data .csv")
sales_detail = pd.read_csv("sales_detail .csv")
sales_summary = pd.read_csv("sales_summary .csv")
```

```python
new_dataset_2017 = pd.read_csv("customer_data _2017.csv", index_col = 0)
new_dataset_2018 = pd.read_csv("customer_data _2018.csv", index_col = 0)
new_dataset_2019 = pd.read_csv("customer_data _2019.csv", index_col = 0)
```

```python
new_dataset_2017.head()
```

```python
sales_summary.columns
```

```python
months = ["", "jan", "feb", "mar", "apr", "may", "jun", "jul" ,"aug", "sep", "oct", "nov", "dec"]
for i in range(sales_summary.shape[0]):
    ID, year, month, month_1 ,month_2,month_3,month_4,end = list(sales_summary.loc[i,:])
    if year == 2017:
        new_dataset_2017.at[ID,months[month] + "_1"] = month_1
        new_dataset_2017.at[ID,months[month] + "_2"] = month_2
        new_dataset_2017.at[ID,months[month] + "_3"] = month_3
        new_dataset_2017.at[ID,months[month] + "_4"] = month_4
    elif year == 2018:
        new_dataset_2018.at[ID,months[month] + "_1"] = month_1
        new_dataset_2018.at[ID,months[month] + "_2"] = month_2
        new_dataset_2018.at[ID,months[month] + "_3"] = month_3
        new_dataset_2018.at[ID,months[month] + "_4"] = month_4
    elif year == 2019:
        new_dataset_2019.at[ID,months[month] + "_1"] = month_1
        new_dataset_2019.at[ID,months[month] + "_2"] = month_2
        new_dataset_2019.at[ID,months[month] + "_3"] = month_3
        new_dataset_2019.at[ID,months[month] + "_4"] = month_4
    
```
